# 1. set

A set is an **unordered** collection of *unique* elements. You can think of them like dicts,
but keys only, no values. A set can be created in two ways: via the set function or via
a set literal with curly braces.<br>
**Note**: to create an empty set you have to use set(), not {}.

````python
# Create a set structure ## vẫn giống khởi tạo 1 list và tuple chỉ khác dùng ngoặc nhọn và chỉ chứa các phần tử duy nhất
x = {1, 2, 3, 3, 3, 4, 5, 5}
print(x)
````

````python
# empty set
empty_set = set()
````

#### set() function to convert

````python
li = [2, 2, 2, 1, 3, 3]
se = set(li)
print(li)
print(se)
````

## 3.4. loop in dictionary

````python
fruit_price = {'apple': 30,
               'banana': 20,
               'cherry': 20,
               'watermelon': 999}
````

### How many ways to loop through dict ? 

````python
# 1
for i in fruit_price:
    print(i)
````

````python
# 2
for i in fruit_price.keys():
    print(i, fruit_price[i])
````

````python
# 3
for i in fruit_price.values():
    print(i)
````

````python
# 4
for k, v in fruit_price.items():
    print(f"{k} => {v}")
````

# 4. List, Set, and Dict Comprehensions
syntax
```
[expression for item in collection if condition] # list comprehension
```

#### list comprehensions

````python
# 4 times for each element in given list
str1 = [1, 2, 3, 4]  # [4, 8, 12, 16]
````

````python
## Normal way 
rs = []
for i in str1:
    rs.append(i*4)
rs
````

````python
## list comprehensions
rs = [i*4 for i in str1]
rs
````

````python
# tạo một list chứa các các giá trị là số chẵn từ list str1
str1 = [1, 2, 3, 4]
rs = []
for i in str1:
    if i % 2 == 0:
        rs.append(i)
rs
````

````python
rs = [i for i in str1 if i % 2 == 0]
rs
````

````python
rs = []

for i in range(5):
    for j in range(3):
        for n in range(2):
            rs.append(i*j*n)
````

````python
# sử dụng list comprehensions nhưng không khuyến khích trong trường hợp này
rs = [i*j*n for i in range(5) for j in range(3) for n in range(2)]
rs
````

#### dict comprehensions

````python
# Create dict : key from 0 - > 9 and value = key^2 using dict comprehension
# {0: 0, 1: 1, 2: 4, 3: 9, ...}
dict1 = {}

for i in range(10):
    dict1[i] = i*i

print(dict1)
````

````python
dict1 = {i:i*i for i in range(10)}
dict1
````

````python
# Create dict: key is odd number from 0->9 and value = key^2
rs = {i:i*i for i in range(10) if i%2 != 0}
rs
````

````python
rs = {}
for i in range(10):
    if i%2 != 0:
        rs[i] = i*i
rs
````
