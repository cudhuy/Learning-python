### Slicing
```string[start:stop:step]``` -> return chuỗi ký tự từ gồm các ký tự từ vị trí start đến vị trí stop - 1 với bước nhảy step

````python
# slicing: string[start:stop]: lay toan bo ki tu tu vi tri start den vi tri stop trong xau
# lay ra tu Datapot
s[0:6] #--> voi slicing, luon dung lai o vi tri so 5 (vi tri stop-1)
# Lay ra tu pot
s[4:7]
````

````python
# Lay ra tu Datapot
print(s[0:7])
print(s[-14:-7])
````

````python
# Lay ra Python bang 2 cach
# Pos:
s[8:14]
# Neg:
s[-6:0] # sai vi duyet tu trai sang phai, khong the lam nguoc lai, 0 o dau xau
s[-6:] # dung
````

#### viết tắt:
 - không khai báo start thì mặc định lấy từ đầu xâu đến stop-1
 - không khái báo stop thì mặc định lấy đến cuối xâu

````python
# Lay ra xau Datapot 3 cach
s[-14:-7]
s[0:7]
s[:7]
s[:-7]
````

````python
print(s[8:14])  # from 8 to 13
print(s[8:])  # from 8 to end
````

````python
print(s[0:7])  # from 0 to 6
print(s[:7])  # from start to 6
print(s[:-7])  # from start to -8
````

# 6. Conditional Statements

Capture d’écran 2020-09-12 à 16.10.28.png

**Syntax:**
```
if <câu điều kiện 1>:
    <câu lệnh thực thi nếu trường hợp điều kiện 1 được thỏa mãn>
elif <câu điều kiện 2>:
    <câu lệnh thực thi nếu trường hợp điều kiện 2 được thỏa mãn>
...
else:
    <câu lệnh thực thi ở các trường hợp còn lại>
```

## Basic if statement

````python
if 2 == 2:  # Check if condition, if True execute if-block (print right)
    print("equal")
    a = 123
    b = a + 1
    print(b) ## 4 dòng này phải thẳng hàng nhau nhé
else:  # if (if condition) return False, then execute else-block (print wrong) ## viết thẳng hàng if nhé
    print('not equal')
    print("akevj") 
````

## Full case with elif, else

````python
x = 0

if x < 0:
    print('negative')
elif x == 0:
    print('zero')
elif 0 <= x < 10:
    print('Positive but smaller than 10')
else:
    print('Greater than 10')
````

## Compare between multiple if and if-elif

````python
n = 20
````

````python
# multiple if
if n > 0:
    print("Number is Positive")  
if n < 0:
    print("Number is Negative")
if n == 0:
    print("Number is ZERO")
````

````python
# if-elif
if n > 0:
    print("Number is Positive")
elif n < 0:
    print("Number is Negative")
else: # == 0
    print("Number is ZERO")
````

## Nested if/else

````python
b = 0
c = 0
````

````python
# Kiểm tra xem b == 0, nếu có thì kiểm tra c == 0, nếu đúng thì in ra 'b = c = 0'
if b == 0:
    if c == 0:
        print('b = c = 0')
````

````python
# if (b == 0) and (c == 0):
if b == c == 0:
    print('b = c = 0')
````
