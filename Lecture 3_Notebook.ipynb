# 2. while loops
A while loop specifies a condition and a block of code that is to be executed *until the condition evaluates to False* or the loop is explicitly ended with *break*.<br>
syntax:
```
while <condition>:
    <codes>
```

Capture d’écran 2020-09-12 à 21.42.53.png

### With the while loop we can execute a set of statements as long as a condition is true

````python
# print number (from 0) until the condition is False (greater than or equal to 10)
count = 0
while count < 10:  # while-condition, if True then execute code in block, else exit while-block
    print(count)
    count = count + 1
````

## 2.1. break statement
The break statement terminates the loop containing it. Control of the program flows to the statement immediately after the body of the loop.

If the break statement is inside a nested loop (loop inside another loop), the break statement will terminate the innermost loop.

11.png

````python
for letter in 'Python':
    if letter == 'h':
        break  # exit loop func when hit h character
    print('Current letter :', letter)
````

````python
var = 10
while var > 0:              
    if var == 5:
        break  # exit when var drops to 5
    print('Current variable value :', var)
    var = var - 1
````

## 2.2. continue
The continue statement is used to skip the rest of the code inside a loop for the current iteration only.<br>Loop does not terminate but continues on with the next iteration.

11.png

````python
for letter in 'Python':
    if letter == 'h':
        continue  # skip print for h character
    print('Current letter:', letter)
````

````python
var = 10
while var > 0:
    var = var - 1
    if var == 5:
        continue
    print('Current variable value :', var)

print('Good bye!')
````

# 3. Data structure - list

A data structure is a data organization, management, and storage format that enables efficient access and modification. 
The basic data structures in Python include <b>list</b>, <b>set</b>, <b>tuples</b>, and <b>dictionary</b>. 

<b>Lists</b> are an important data type that allows you to keep a set of iterable item(s). Lists are *MUTABLE* sequences. This literally means you can have a list of items, like 1, 2, 3 or "h", "e", "l", "l", "o", "!".<br> A list is denoted by square brackets, [ ]. To create a list that holds actual items, put them within square brackets and use commas to separate the items.

## 3.1. Create a list

````python
lst = [1, 2, 3, 4]
print(lst)
````

````python
# multiple diffenrent data types
lst = [1, 2.2, 'F', 1+2j, False]
````

````python
# nested list
lst = [['h', 'e', 'l', 'l', 'o'], 
        [True, 13, None, 'l', 'd', '123', 123]]
````

### List indexing/slicing

````python
team = ['linh', 'chi', 'huong', 'hien', 'trang']
print("First item: ", team[0])
print("List of first two items:", team[0:2])
````

## 3.2. built-in list methods

#### list.append() : add AN ITEM to the end of the list

````python
team = ['linh', 'chi', 'huong', 'hien', 'trang']
team.append('bao')
team
````

````python
team.append(['bao', 'viet'])
team
````

#### list.extend(): extend LIST by adding all the items from iterable

````python
team = ['linh', 'chi', 'huong', 'hien', 'trang']
team.extend(['bao', 'viet'])
team
````

#### list.insert(i, x): insert an item x at a given position i

````python
team = ['linh', 'chi', 'huong', 'hien', 'trang']
print("Before", team)
team.insert(3, 'trang')
print("After", team)
````

#### list.remove(x): remove the **first** item from the list whose value is equal to x

````python
test = ['trang', 'linh', 'chi', 'chi']
test.remove('chi')
test
````

#### list.pop([i]): remove the item at the given position in the list and return it

````python
team
````

````python
team.pop(0)  # remove the last item
team
````

#### list.count(x): return the number of times x appears in the list

````python
print(team)
occu_trang = team.count('trang')
print(f"Trang xuat hien {occu_trang} lan trong list")
````

# 4. Tuple
In contrast with list , a tuple is a fixed-length, **IMMUTABLE** sequence of Python objects. The easiest way to create one is with a comma-separated sequence of values: (1, 2, 3).

## 4.1 Create a tuple

````python
tup = (2, 3, 5)
````

````python
tup_2 = (1, 'Hello', True, False, None)
````

#### nested tuple

````python
nested_tup = ((4, 5, 6), 
              (7, 8, 9),
              [10, 11, 12])
````

#### Convert a non-tuple to a tuple

#### Convert a list to a tuple

````python
li = [3, 5, 6]
a = tuple(li)
a
````

#### Convert a string to a tuple

````python
tuple('a string')
````

#### Access Tuple Items: indexing/slicing

````python
tup = (2, 3 ,5)
tup[:2]
````

## 2.2 Tuple is immutable
tuple is immutable that means you can not change the value of tuple 

````python
tup = ('123', [1, 2], True)
tup[2] = 123
tup
````

#### So why do we need tuple even if this data structure isn't convenient as list ?

## 2.3 Unpacking a tuple/list

````python
lst = [1, 2, 3]
x, y, z = lst
````

````python
x, y, z = [1, 2, 3]
z
````

````python
tup = (6, 8, 9)
a, b, c = tup
print(a, b, c)
````

````python
pair = ('phone', [123, 234])
key, [value1, value2] = pair
print(key, value1, value2)
````
