## Task 6
Viết hàm kiểm tra 1 list đầu vào có chứa các giá trị 1, 2, 3 liền kế đúng thứ tự hay không.<br>
Ví dụ
- contains_123(['a', 'b', 'c']) -> Không
- contains_123(['a', 1, 2, 'b', 3]) -> Không
- contains_123(['1', 1, 'a', 'b', 1, 2, 3, 'c']) -> Có

````python
def contains_123(input_list: list):
    # Start here
    for i in range(len(input_list) - 2):
            if input_list[i] == 1 and input_list[i+1] == 2 and input_list[i+2] == 3:
                return "Có"
    return "Không"
````

````python
# Huy test
print(contains_123(['a', 'b', 'c']))
print(contains_123(['a', 1, 2, 'b', 3]))  
print(contains_123(['1', 1, 'a', 'b', 1, 2, 3, 'c']))
````
